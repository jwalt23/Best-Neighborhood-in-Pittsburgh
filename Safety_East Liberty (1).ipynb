# Safety Analysis of East Liberty

### Comparing East Liberty with Selected Pittsburgh Neighborhoods

## Introduction

This notebook analyzes safety in East Liberty by comparing it with several selected Pittsburgh neighborhoods: Central Business District, South Side Flats, and Carrick. 

I selected these neighborhoods because they are well-known areas with notable crime activity, which makes the comparison useful for understanding how East Liberty performs in terms of safety.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("044f2016-1dfd-4ab0-bc1e-065da05fca2e (1).csv")
df.head()
```

## Exploring the Dataset

First, I loaded the Pittsburgh crime dataset and reviewed its structure. This helps identify the column used to compare crime incidents by neighborhood.

```python
df.columns
```

```python
crime_by_neighborhood = df["INCIDENTNEIGHBORHOOD"].value_counts()
crime_by_neighborhood.head(10)
```

## The Metric

My sub-metric is safety. I measured safety by counting the number of crime incidents in each selected neighborhood. A lower crime count indicates a safer neighborhood.

For this analysis, I focused on East Liberty and compared it with a few selected Pittsburgh neighborhoods rather than every neighborhood in the city.

```python
east_liberty = df[df["INCIDENTNEIGHBORHOOD"] == "East Liberty"]
east_liberty_count = east_liberty.shape[0]

print("Number of crime incidents in East Liberty:", east_liberty_count)
```

## Selected Neighborhood Comparison

To make the comparison clear, I selected the following neighborhoods:

- East Liberty
- Central Business District
- South Side Flats
- Carrick

```python
selected = crime_by_neighborhood[
    ["East Liberty", "Central Business District", "South Side Flats", "Carrick"]
]

selected
```

```python
selected.plot(kind="bar", figsize=(8,5))
plt.title("Safety Comparison Between Selected Neighborhoods")
plt.xlabel("Neighborhood")
plt.ylabel("Crime Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Results

East Liberty had 8451 crime incidents. Compared to the selected neighborhoods, it had fewer incidents than Central Business District, South Side Flats, and Carrick.

Based on this comparison, East Liberty appears to be safer than the other selected neighborhoods in this analysis.
