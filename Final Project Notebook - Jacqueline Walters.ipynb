Introduction:
Our project focuses on evaluating datasets to determine which neighborhood can be considered the “best” based on measurable, data‑driven criteria. Because “best” is subjective, our first step was to identify which qualities matter most to residents and how those qualities could be quantified in a meaningful way.
I approached this by exploring several possible dimensions of neighborhood quality. Initially, I considered metrics related to recreation—such as access to playgrounds, sports facilities, and community centers—as well as indicators of family‑friendliness, like school proximity or childcare availability. These were strong candidates because they directly influence daily life and community engagement.
However, as I compared these options, I realized that many of them overlapped or were difficult to measure consistently across neighborhoods. That led me to explore a different angle: eco‑friendliness, specifically the amount of green space available in each neighborhood.


The Metric:
The metric I developed focuses on eco‑friendliness, which I define as the degree to which a neighborhood provides access to nature, green space, and environmentally supportive features. The goal is to capture how well each neighborhood supports outdoor activity, environmental quality, and overall sustainability.
The Eco‑Friendliness Score is built from features that directly reflect the presence and accessibility of natural spaces. These features are quantifiable, comparable across neighborhoods, and grounded in the datasets available.
https://data.wprdc.org/dataset/operations-green-spaces (My Dataset)
https://data.wprdc.org/dataset/operations-green-spaces/resource/b06fa202-ead7-491c-b8f6-2702dd35f8dd

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("5cf7163b-f76c-47ef-b48d-a16366ac1835.csv")
df = df[['Neighborhood', "Total Acreage", "Maintainable Acreage"]]

df['Total Acreage'] = pd.to_numeric(df['Total Acreage'], errors='coerce')
df['Maintainable Acreage'] = pd.to_numeric(df['Maintainable Acreage'], errors='coerce')


df = df.dropna(subset=['Neighborhood'])


metric = 'Total Acreage'   


grouped = df.groupby('Neighborhood')[metric].sum().sort_values()

print(grouped)
print("\nNeighborhood with the lowest total acreage:", grouped.idxmin())

# Plot
plt.figure(figsize=(12,6))
grouped.plot(kind='bar')
plt.title(f"Total {metric} by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel(metric)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
