Our project focuses on evaluating datasets to determine which neighborhood can be considered the “best” based on measurable, data‑driven criteria. Because “best” is subjective, our first step was to identify which qualities matter most to residents and how those qualities could be quantified in a meaningful way. Our submetrics are eco-friendliness, safety, cleanliness and toxicity levels. We also explored affordability, walkability, property-level quality, and school quality but decided our final choices were the best. Our submetrics are accurate measurements of what the best neighborhood is because our selection is diverse and covers a lot of areas. All these factors contribute to the quality of life in a neighborhood which is important when considering which one is the best.

The Metric: Describe your metric, and what features are you measuring. What datasets are you using?

Our metric evaluates neighborhoods across four main features:
1. Eco‑Friendliness
Measured using the Operations – Green Spaces dataset from the Western Pennsylvania Regional Data Center. We analyzed:
- Total acreage of green spaces
- Maintainable acreage
- Number and distribution of green‑space facilities
This allowed us to quantify how much natural space each neighborhood offers.
2. Safety
Measured using crime‑related datasets (e.g., incident reports). We aggregated incidents by neighborhood to estimate relative safety levels.
3. Cleanliness
Measured using datasets related to public works, maintenance, or 311 cleanliness‑related requests. This helped us approximate how well‑maintained each neighborhood is. Shadyside is objectively the best neighborhood in Pittsburgh. It ranks #1 out of 79 neighborhoods, sits 11% ahead of East Liberty (the runner-up), and scores 5x above the citywide mean. Its investment is consistent year-over-year and reflects the surrounding environment.
4. Toxicity Levels
Measured using the Toxic Release Inventory dataset. We filtered by ZIP code, mapped ZIPs to neighborhoods, and summed total toxic releases to determine which neighborhoods experience the least environmental pollution.
Each submetric was normalized and compared across neighborhoods to identify which area consistently performed well across categories.
